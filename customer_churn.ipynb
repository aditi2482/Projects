# Removing Skewness

```python
df.skew()
```

```python
skew=('TotalCharges','SeniorCitizen')
for col in skew:
    if df.skew().loc[col]>0.55:
        df[col]=np.log1p(df[col])
```

```python
df = df.drop(['customerID'], axis=1)    
```

# Label Encoding

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

df['gender']= label_encoder.fit_transform(df['gender'])
df['SeniorCitizen']= label_encoder.fit_transform(df['SeniorCitizen'])
df['Partner']= label_encoder.fit_transform(df['Partner'])
df['Dependents']= label_encoder.fit_transform(df['Dependents'])
df['PhoneService']= label_encoder.fit_transform(df['PhoneService'])
df['MultipleLines']= label_encoder.fit_transform(df['MultipleLines'])
df['InternetService']= label_encoder.fit_transform(df['InternetService'])
df['OnlineSecurity']= label_encoder.fit_transform(df['OnlineSecurity'])
df['OnlineBackup']= label_encoder.fit_transform(df['OnlineBackup'])
df['DeviceProtection']= label_encoder.fit_transform(df['DeviceProtection'])
df['TechSupport']= label_encoder.fit_transform(df['TechSupport'])
df['StreamingTV']= label_encoder.fit_transform(df['StreamingTV'])
df['StreamingMovies']= label_encoder.fit_transform(df['StreamingMovies'])
df['Contract']= label_encoder.fit_transform(df['Contract'])
df['PaperlessBilling']= label_encoder.fit_transform(df['PaperlessBilling'])
df['PaymentMethod']= label_encoder.fit_transform(df['PaymentMethod'])
df['Churn']= label_encoder.fit_transform(df['Churn'])
```

# Model Building & Evaluation

```python
x=df.iloc[:,:19]
y=df['Churn']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
features = x.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(x)
x= pd.DataFrame(scaler.transform(x))
x.columns = features
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=41)
```

```python
x_train = x_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
x_test = x_test.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score,roc_curve
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
prediction_logreg=logreg.predict(x_test)
print(accuracy_score(y_test,prediction_logreg))
```

# Random Forest Classifier with Randomized Search CV

```python
from sklearn.ensemble import RandomForestClassifier
rf_c=RandomForestClassifier()
```

```python
param_grid={'n_estimators':[int(x) for x in np.linspace(start=200,stop=1200,num=11)],
            'max_features':['auto','sqrt'],
            'max_depth':[int(x) for x in np.linspace(start=10,stop=100,num=11)],
            'min_samples_leaf':[1,2,3,5],
            'min_samples_split':[2,5,10,15]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_cv=RandomizedSearchCV(rf_c,param_grid,cv=3,verbose=2,random_state=42)
random_cv.fit(x_train,y_train)
best_random=random_cv.best_estimator_
prediction_cv=best_random.predict(x_test)
print(accuracy_score(y_test,prediction_cv))
```

# AdaBoost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
abc_model = AdaBoostClassifier()
abc_model.fit(x_train, y_train)
prediction_abc = abc_model.predict(x_test)
print(accuracy_score(y_test, prediction_abc))
```

# Confusion Matrix

```python
print(confusion_matrix(y_test,prediction_abc))    
```

# ROC Curve

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
model= AdaBoostClassifier()
model= model.fit(x_train,y_train)
pred_prob=model.predict_proba(x_test)[:,1]
pred_prob
```

```python
fpr,tpr,thresholds=roc_curve(y_test,pred_prob,pos_label=True)
```

```python
print(fpr)
print("/n")
print(tpr)
print("/n")
print(thresholds)
```

```python
plt.plot([0,1],[1,0],'d--')
plt.plot(fpr,tpr,label='ROC Curve')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC Curve')
plt.show()
```

```python
import pickle 
filename="churn.pkl"
pickle.dump(model,open(filename,'wb'))
```
